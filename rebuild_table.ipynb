```python
from selenium import webdriver
```

```python
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
```

```python
driver=webdriver.Chrome(ChromeDriverManager().install())
```

```python
driver.get("https://www.nseindia.com/api/option-chain-indices?symbol=NIFTY")
data=driver.find_element(By.XPATH,"//pre")
data.text
```

```python
import json
json_data=json.loads(data.text)
print(json.dumps(json_data,indent=2))
```

```python
records=json_data["records"]["data"]
```

```python
for item in records:
    if item['expiryDate']=='28-Jul-2022':
        print(item)
```

```python
lis=['call_oi','call_chng_in_oi','call_volume','call_iv','call_ltp','call_chng','call_bid_qty','call_bid_price','call_ask_price','call_ask_qty','strike price','put_bid_qty','put_bid_price','put_ask_price','put_ask_qty','put_chng','put_ltp','put_iv','put_volume','put_chng_in_oi','put_oi']
scrap=["openInterest","changeinOpenInterest","totalTradedVolume","impliedVolatility","lastPrice","change","bidQty","bidprice","askPrice","askQty","strikePrice","bidQty","bidprice","askPrice","askQty","change","lastPrice","impliedVolatility","totalTradedVolume","changeinOpenInterest","openInterest"]
dic={}
for val in lis:
    dic[val]=[]
print(len(lis),len(scrap))
```

```python
for item in records:
    if item['expiryDate']=='28-Jul-2022':
        for i in range(len(lis)):
            if i<=10:
                dic[lis[i]].append(item['CE'][scrap[i]])
            else:
                dic[lis[i]].append(item['PE'][scrap[i]])
        
```

```python
import pandas as pd
df=pd.DataFrame(dic)
df.head()
```

```python
df.dtypes
```

```python
df=df.replace('-','0')
```

```python
call_oi=df['call_oi'].sum()
call_volume=df['call_volume'].sum()
put_oi=df['put_oi'].sum()
put_volume=df['put_volume'].sum()
print(call_oi,call_volume,put_oi,put_volume)
df=df.replace(0,'-')
df=df.replace(0.0,'-')
```
